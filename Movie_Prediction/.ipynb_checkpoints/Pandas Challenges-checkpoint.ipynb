*Plot domestic total gross over time*

```python
df.sort('ReleaseDate')

plt.figure(figsize=(15,10))
plt.plot_date(df.ReleaseDate, df.DomesticTotalGross)
plt.xlabel('Release date')
plt.ylabel('Domestic total gross')
plt.title('Domestic total gross over time')
```

*Plot runtime vs. domestic total gross*

```python
plt.figure(figsize=(15,10))
plt.plot_date(df.Runtime, df.DomesticTotalGross)
plt.xlabel('Runtime')
plt.ylabel('Domestic total gross')
plt.title('Domestic total gross by runtime')
```

Bin the dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean.

```python
#change index in df to datetime object to organize by month
df2 = df.set_index('ReleaseDate')
```

```python
#organize Domestic Total Gross by mean per month. 

#df3
df3 = df2.resample("M", how='mean')
df3 = df3.rename(index=lambda x: x.strftime('%B'))
df3
#df3.groupby([df.ReleaseDate.index.month]).first()
```

```python
#Group dates by month, find the Standard Deviation and Mean of Domestic Total Gross
df4 = df2.resample('M').agg(['mean','std'])
df4 = df4.rename(index=lambda x: x.strftime('%B'))
df4
```

```python
# create mean and STD columns, rename columns
df4.columns = [(col) for col in df4.columns]
df4
df4.columns = ['Budget_mean',
               'Budget_STD',
               "DomesticTotalGross_mean",
               "DomesticTotalGross_STD",
               "Runtime_mean",
               "Runtime_STD"
]
df4
```

```python
#plot the Mean domestic total gross by month in 2013, add the standard deviation column


ax = df4[['DomesticTotalGross_mean']].plot(kind='bar', yerr=df4["DomesticTotalGross_STD"],
                                      title ="Mean domestic total gross by month in 2013",
                                      figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Mean domestic total growth", fontsize=12)
plt.show()
```
