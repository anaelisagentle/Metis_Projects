### Run a regression to test features with imdbRating as target

Already, there is a high R-squared value with a high Adj. R-squared. A high R-squared tells us that the data is well-fit to the regression line. The high adjusted R-squared shows that regression hasn't been overfit. The Omnibus results below show a p-value of 0.00. This means we reject the null and assume there is skew in the data. It's possible that a polynomial regression should be used. 

```python
y = df_test.imdbRating
#x = df_test.drop('imdbRating',1)
x = df_test.drop(['imdbRating',"rating[T.Unknown]","Intercept", "Runtime"],1)

lsm = sm.OLS(y,x)
fit4 = lsm.fit()
fit4.summary()
```

```python
sns.pairplot(df_test, size = 1.2, aspect=1.5)
```

The residual plot below should be completely random, but there is a trend around 0 on the y-axis

```python
fit4.resid.plot(style='o', figsize=(12,8))
```

```python
#test for Heteroscedasticity (sp?) with Breush-Pagan test
name = ['Lagrange multiplier statistic', 'p-value',
        'f-value', 'f p-value']
test = sms.het_breushpagan(fit4.resid, fit4.model.exog)
lzip(name, test)
```

A Heteroscedasticity test is run and new polynomial values are passed into the model. The R-squared drops down, so this is not a good model

```python
pf = pp.PolynomialFeatures(degree = 2)
X = pf.fit_transform(x)
```

```python
lsm = sm.OLS(y,X)
fit5 = lsm.fit()
fit5.summary()
```

```python
# 10-fold cross-validation
reg = LinearRegression()
scores = cross_val_score(reg, x, y, cv=10, scoring='mean_squared_error')

# scores output is negative, a sklearn quirk bc mse is used to min. optimization func.
mean_sq = (-scores)
print mean_sq
```
